```python
import matplotlib.pyplot as plt 
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('alla.csv')
df.head()
```

```python
print("Total of {} entries in the dataset before cleaning.".format(df.shape[0]))
```

```python
# We get rid of website html numeric listing:
```

```python
df = df.drop(columns=['Unnamed: 0']) 
df
```

```python
#now lets try to get rid of duplications
df.drop_duplicates(keep='first')
df
```

```python
#there are no duplications in the data
```

```python
#Let's get rid of the NaN values
```

```python
df = df.dropna()
print("Total of {} entries in the dataset after cleaning Nulls.".format(df.shape[0]))
```

```python
df
```

```python
#Here we get rid of unnecessary spaces and chars.
#Also, we did a casting from str to int
```

```python
#link to solution from StackOverflow: https://stackoverflow.com/questions/20625582/how-to-deal-with-settingwithcopywarning-in-pandas

df['num_of_users'] = df['num_of_users'].str.replace(' Ratings','')
df = df.astype({"num_of_critics":"int","critics_grade":"int","release_year":"int","num_of_users":"int"})
df[['num_of_critics', 'critics_grade', 'user_grade', 'release_year','num_of_users']].describe(include='all')
```

```python
#Here is the correlation table:
```

```python
df.corr()
```

```python
sns.heatmap(df.corr());
```

```python
#as we can see there is a correlation betwwen user grades and critic grades
```

```python
fig = plt.figure(figsize=(25,10))
plt.scatter(df['game_developer'], df['critics_grade'])
plt.title('Game Developers Versus Critics Grade')
plt.xlabel('Game Developer')
plt.xticks(rotation=90)
plt.xticks(fontsize=5)
plt.ylabel('Critics Grade')
plt.show()
```

```python
fig = plt.figure(figsize=(25,10))
plt.scatter(df['game_developer'], df['user_grade'])
plt.title('Game Developers Versus Users Grade')
plt.xlabel('Game Developer')
plt.xticks(rotation=90)
plt.xticks(fontsize=5)
plt.ylabel('Users Grade')
plt.show()
```

```python
#as we can see, the developers play an important role in the user score and critics score 
#but we cant use it for prediction of popularity becuse we nee more info about it
```

```python
df['platform'] = df['platform'].str.replace(' ','')
plt.scatter(df['platform'], df['critics_grade']);
plt.title('Platform Versus Critics Grade');
plt.xlabel('Platform');
plt.xticks(rotation=90)
plt.ylabel('critics grade');
```

```python
#here we can see how the critics grades scatter between/across platforms^
```

```python
plt.scatter(df['platform'], df['user_grade']);
plt.title('Platform Versus Users Grade');
plt.xlabel('Platform');
plt.xticks(rotation=90)
plt.ylabel('Users Grade');
```

```python
#here we can see how the user grades scatter between platforms^

```

```python
#let's explore the top the frequent developer:
```

```python

counts = df['game_developer'].value_counts()

fig, ax = plt.subplots(figsize=(12,6))
top_10 = counts[:10] 
plt.bar(range(10), top_10.values, align='center', tick_label=top_10.keys());
plt.title("Top ten frequent developers");
plt.xticks(rotation=90)
```

```python
#here we divided the critics grade by 10 for comparison reasons:
```

```python
fig = plt.figure(figsize=(10,10))
ax = plt.axes()
ax.scatter(df.critics_grade/10, df.user_grade)
plt.xlabel('Game Critics grade')
plt.ylabel('Game User grade')
plt.show()
```
